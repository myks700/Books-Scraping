```python
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import time
import warnings
warnings.filterwarnings('ignore')
import time
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://www.gfxtra31.com/programming-books/')
```

```python
books_url = []
Name=[]
URL = []
start = 0
end = 5

for page in range(start, end):
    url = driver.find_elements(By.XPATH, '//div[@class="baslik"]/a')
    for i in url:
        books_url.append(i.get_attribute('href'))
    nxt_btn = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/div[1]/div[1]/div[1]/div/div[25]/a[11]')
    nxt_btn.click()
    time.sleep(2)   
    
    
    
```

```python
len(books_url)
```

```python
for url in books_url:
    driver.get(url)
    time.sleep(2)
   
    try:
        name = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/div[1]/div[1]/div[1]/div/div[1]/div/div[1]/div[1]/a')
        Name.append(name.text)
        url = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/div[1]/div[1]/div[1]/div/div[1]/div/div[1]/div[1]/a')
        URL.append(url.get_attribute('href'))
        
        
    except NoSuchElementException:
        Name.append('_')
```

```python
name = pd.DataFrame(Name, columns = ['Product_Name']).head(55)
url =  pd.DataFrame(URL, columns = ['Product_url']).head(55)
```

```python
books_data = pd.concat([name, url], axis = 1)
books_data
```

```python
books_data.to_excel("books_data.xlsx") 
```
